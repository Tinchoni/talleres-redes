```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
red1 = pd.read_csv("tabla_red1.csv",index_col=0)
red2 = pd.read_csv("tabla_red2.csv",index_col=0)


x = red2.index
entropia_red1 = list(red1['Entropia'])
paquetes_red1 = list(red1['Paquetes'])
entropia_red2 = list(red2['Entropia'])
```

```python
sns.set_style("darkgrid")

plt.title("Entropia por hora")
plt.xticks(rotation=65)
plt.plot(x,entropia_red1)
plt.plot(x,entropia_red2)
plt.legend(['Red1', 'Red2'])
```

```python
sns.set_style("darkgrid")

plt.title("Paquetes por hora")
plt.xticks(rotation=65)
plt.plot(x,paquetes_red1)
plt.legend(['Red1'])
```
